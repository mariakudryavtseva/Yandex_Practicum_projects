## Предобработка данных

### Обработка пропусков

Найдем количество пропусков и их процент в нашем наборе данных с объявлениями о продаже квартир.

```python
# считаем количество пропусков
df_na_count = df.isna().sum()
# считаем процент пропусков
percent_nan = (df_na_count * 100 / len(df)).round(2)
# создание датафрейма с результатом
pd.DataFrame({'count': df_na_count,
              'percent_missing (%)': percent_nan}
            ).sort_values(by='count', ascending=False)
```

Заметим небольшое количество пропусков в столбце `locality_name`, который содержит название населенного пункта, в котором находится объект недвижимости. Поскольку нам неизвестен населенный пункт для таких объявлений, мы не можем использовать их в исследовании. Таким образом, допустимо удаление этих строк в таблице с данными.

```python
# удаление строк, содержащих Nan в столбце
# locality_name
df = df.dropna(subset=['locality_name'])
```

Столбец `is_apartment` содержит большое количество пропусков и можно предположить, что если не было указано является ли объект недвижимости апартаментами, то подразумевается, что не является. Поэтому заменим все пропуски на значение `False`.

```python
# замена всех Nan в столбце is_apartment на False
df['is_apartment'] = df['is_apartment'].fillna(False)
# проверка уникальных значений столбца is_apartment
df['is_apartment'].value_counts()
```

Согласно пояснению к данным, апартаменты — это нежилые помещения, не относящиеся к жилому фонду. Для нашего исследования такие объявления не представляют интереса, поэтому можем исключить строки таблицы, у которых значение в столбце `is_apartment` равно `True`.

```python
# перезаписываем в таблицу столбец is_apartment,
# значения которого равны только False
df = df[df['is_apartment'] == False]
# проверка уникальных значений столбца is_apartment
df['is_apartment'].value_counts()
```

Большое количество пропусков в столбце `balcony` можно трактовать как отсутствие балкона в объекте недвижимости. Поэтому заменим такие пропуски на нулевые значения.

```python
# замена всех Nan в столбце balcony на ноль
df['balcony'] = df['balcony'].fillna(0)
```

Пропуски в столбце `kitchen_area` могут означать, что объект недвижимости имеет открытую планировку или является студией без выделенной кухни. Кроме того, этот параметр могли не указать при размещении объявления. Поэтому оставим эти значения пустыми, а не нулевыми, так как можно предположить, что в таких объектах кухня все-таки присутствует или в отдельном помещении, но не указана площадь, или совмещена с другой комнатой.

Пропуски в столбце `floors_total` оставим без изменений, потому что нет возможности предположить или рассчитать количество этажей в доме, в котором продается квартира. 

Столбцы `airports_nearest`, `cityCenters_nearest`, `parks_nearest`, `ponds_nearest` содержат в себе расстояния до аэропорта, центра города, парка и водоема соответственно. Пропуски в них могут означать, что парков и водоемов нет поблизости или что расстояние до этих мест неизвестно (отслеживается также по пропускам и нулевым значениям в соотвествующих столбцах `parks_around3000` и `ponds_around3000`). Мы не можем заменить их нулевыми значениями, поскольку это расстояние, поэтому оставим эти пропуски пустыми.

Пропуски в столбцах `parks_around3000` и `ponds_around3000` можно трактовать как отсутствие парков и водоемов поблизости (что подтверждается пропусками в значениях расстояния до них) или как отсутствие информации об их наличии рядом с объектом недвижимости. Такие пропуски заменим нулевыми значениями.

```python
# замена всех Nan в столбце parks_around3000 на ноль
df['parks_around3000'] = df['parks_around3000'].fillna(0)
# замена всех Nan в столбце ponds_around3000 на ноль
df['ponds_around3000'] = df['ponds_around3000'].fillna(0)
```

Столбец таблицы `room`, описывающий количество комнат в квартире, не содержит пропусков, но проверим какие значения в нем содержатся.

```python
# # проверка уникальных значений столбца rooms
df['rooms'].value_counts()
```

Замечаем, что для некоторых объектов указано нулевое количество комнат. Возможно, это ошибка заполнения объявления для квартиры типа "студии" или имеется в виду, что изолированных комнат в квартире нет. Оценив такие объявления можно заметить, что наибольшая часть из них имеет небольшую площадь (до 40 м²). Объединим эти значения с однокомнатными квартирами.

```python
# замена нулевого значения столбца rooms на единицу 
df['rooms'] = df['rooms'].replace(0, 1)
```

Пропуски в столбце `ceiling_height` заменим медианным значением высоты потолков по другим объектам недвижимости в виду того, что высота потолков имеет небольшое колебание.

```python
# замена пропусков в столбце ceiling_height на
# медианное значение по этому столбцу
df['ceiling_height'] = df['ceiling_height'].fillna(df['ceiling_height'].median())
```

Пропуски в столбце `days_exposition` заменим медианным значением количества дней размещения объявления для других квартир.

```python
# замена пропусков в столбце days_exposition на
# медианное значение по этому столбцу
df['days_exposition'] = df['days_exposition'].fillna(df['days_exposition'].median())
```

Поскольку пропусков в столбце таблицы с общей площадью квартиры `total_area`, нет, то для столбца с пропусками `living_area` мы можем рассчитать медианный коэффициент отношения жилой площади к общей по имеющимся данным и использовать его для заполнения пропусков в столбце `living_area`. Пропуски в этих данных можно объяснить тем, что очень часто такая информация неизвестна собственникам и в объявлении указывается только общая площадь объекта.

```python
# расчет медианного коэфициента отношения значений living_area
# к значениям total_area
area_ratio = (df['living_area'] / df['total_area']).median()
# замена пропусков в столбце living_area на произведение
# значения total_area на медианный коэффициент
df['living_area'] = df['living_area'].fillna(df['total_area'] * area_ratio)
```

Проверим данные в таблице после обработки пропусков.

```python
df.isna().sum() # подсчёт пропусков
```

### Приведение типов данных

Обратим внимание на то, что данные в некоторых колонках нужно привести к подходящему типу. В колонках `last_price`, `floors_total`, `balcony`, `parks_around3000`, `ponds_around3000`, `days_exposition`, `airports_nearest`, `cityCenters_nearest`, `parks_nearest` и `ponds_nearest` данные имеют тип `float64`. Приведем данные в этих столбцах к целоцисленному типу `int64` в виду того, что:
* для цены недвижимости в колонке `last_price` мы можем отбросить дробную часть, определяющую копейки
* для количества этажей `floors_total` и числа балконов `balcony` дробная часть в значении не имеет смысла
* так как расстояние в столбцах `airports_nearest`, `cityCenters_nearest`, `parks_nearest` и `ponds_nearest` представлено в метрах, не имеет значения дробная часть, указывающая на сантиметры
* количество парков и водоемов (`parks_around3000`, `ponds_around3000`) должно быть целым числом по смыслу, а также продолжительность размещения объявления в днях (`days_exposition`) нас интересует в целых днях, а не в долях

```python
# изменение типа данных float на int,
# для столбцов с Nan меняем float на Int
df = df.astype({
    'last_price': 'int64',
    'floors_total': 'Int8',
    'balcony': 'int8',
    'parks_around3000': 'int8',
    'ponds_around3000': 'int8',
    'days_exposition': 'int32',
    'airports_nearest': 'Int32',
    'cityCenters_nearest': 'Int32',
    'parks_nearest': 'Int32',
    'ponds_nearest': 'Int32'
               })
df.info()
```

Столбец `first_day_exposition`, содержащий дату публикации объявления имеет тип `object`, но для дальнейшей работы с этим столбцом нам необходимо привести его к типу `datetime64`.

```python
# изменение типа данных object на datetime
df['first_day_exposition'] = pd.to_datetime(df['first_day_exposition'], format='%Y-%m-%dT%H:%M:%S')
```

```python
# проверка изменений в общей информации о таблице df
df.info()
```

Методом `astype()` была произведена замена типа данных в столбцах `last_price`, `floors_total`, `balcony`, `parks_around3000`, `ponds_around3000`, `days_exposition`, `airports_nearest`, `cityCenters_nearest`, `parks_nearest` и `ponds_nearest` с вещественного на целочисленный, так как дробная часть для этих данных или не имеет смысла, или не имеет существенного значения для нашего исследования.
Столбец `first_day_exposition`, содержащий дату публикации объявления, был приведен к типу `datetime64`.

## Расчёты и добавление результатов в таблицу

### Расчет цены квадратного метра квартиры

Расчитаем цену квадратного метра для каждого объявления как отношение конечной цены объекта к его общей площади и поместим результаты в новый столбец `price_per_sqm`.

```python
# расчет цены кв. метра и приведение результата к типу int
df['price_per_sqm'] = (df['last_price'] / df['total_area']).astype(int)
```

### Расчет дня недели, месяца и года публикации объявления

Создадим дополнительные столбцы `exposition_day`, `exposition_month` и `exposition_year` для хранения дня, месяца и года публикации объявления.

```python
# создание столбца с днем недели публикации объявления
df['exposition_day'] = df['first_day_exposition'].dt.day_name()
# создание столбца с месяцем публикации объявления
df['exposition_month'] = df['first_day_exposition'].dt.month_name()
# создание столбца с годом публикации объявления
df['exposition_year'] = df['first_day_exposition'].dt.year
```

### Категоризация этажа квартиры

Объединим данные об этажах, на которых находятся объекты недвижимости, в следующие категории:
* первый
* последний
* другой

```python
# функция определения категории,
# принимает данные из столбцов floor и 
# floors_total, возвращает название категории
def floor_categorize(row):
        floor = row[0]
        total = row[1]
        if floor == 1:
            return 'первый'
        if floor == total:
            return 'последний'
        return 'другой'
# применение функции определения категории,
# сохранение категории в столбце floor_category
df['floor_category'] = df[['floor', 'floors_total']].apply(floor_categorize, axis=1)
```

### Расчет соотношения жилой и общей площади квартиры

```python
# результат отношения значений из столбцов
# living_area и total_area помещаем в 
# новый столбец living_ratio
df['living_ratio'] = df['living_area'] / df['total_area']
```

### Расчет соотношения площади кухни к общей площади квартиры

```python
# результат отношения значений из столбцов
# kitchen_area и total_area помещаем в 
# новый столбец kitchen_ratio
df['kitchen_ratio'] = df['kitchen_area'] / df['total_area']
```

### Факторы, влияющие на стоимость квартиры

#### Зависимость цены от площади квартиры

Можно предположить, что цена квартиры, несомненно, зависит от ее площади. Определим характер этой зависимости. Построим график и оценим значения.

```python
# построение графика зависимости цены квартиры
# от ее площади
df.plot(
    x='total_area',
    y='last_price',
    style='o',
    grid=True,
    figsize=(12, 5),
    title='Зависимость цены от площади'
    );
```

Значений цены для каждой площади очень много и в данном виде график сложно оценить. Попробуем использовать группировку по площади квартиры и построить график по медианным значениям цены.

```python
# группируем данные по total_area,
# выбираем медианное значение last_price 
# и строим график их зависимости
df.pivot_table(index='total_area', 
               values='last_price', 
               aggfunc='median'
              ).plot(grid=True, figsize=(12, 5), title='Зависимость медианной цены от площади');
```

По данному графику можно сказать, что цена объекта недвижимости растет пропорционально площади объекта, но после значений равных 100 кв.м. этот рост замедляется. Вероятно, это связано с уменьшением цены за квадратный метр для объектов такой площади. Проверим эту гипотезу и построим график зависимости площади квартиры от медианной цены за квадратный метр.

```python
# группируем данные по total_area,
# выбираем медианное значение price_per_sqm 
# и строим график их зависимости
df.pivot_table(index='total_area', 
               values='price_per_sqm', 
               aggfunc='median'
              ).plot(grid=True, 
                     figsize=(12, 5), 
                     title='Зависимость медианной цены за кв. м. от площади');
```

Действительно, на графике видим, что рост цены за квадратный метр снижается для жилья более 100 кв.м. 

На линейную взаимосвязь между площадью квартиры и ее ценой может также указать коэффициент корреляции Пирсона. Рассчитаем его.

```python
# вычисление и вывод коэффициента корреляции
print(df['total_area'].corr(df['last_price']))
```

Коэффициент корреляции положительный и близок к единице, что говорит о том, что линейная взаимосвязь достаточно тесная и с ростом одной величины будет расти вторая.

```python
# группируем данные по rooms,
# выбираем медианное значение last_price 
# и строим график их зависимости
df.pivot_table(index='rooms', 
               values='last_price', 
               aggfunc='median'
              ).plot(grid=True, 
                     figsize=(12, 5), 
                     title='Зависимость медианной цены от числа комнат');
```

Очевидно, что цена растет с ростом количества комнат. Рассчитаем также коэффициент корреляции Пирсона и убедимся в линейной зависимости цены от количества комнат.

```python
# вычисление и вывод коэффициента корреляции
print(df['rooms'].corr(df['last_price']))
```

Коэффициент корреляции положительный, но не так близок к единице, как в зависимости от общей площади квартиры, линейная связь не такая сильная. Тем не менее это доказывает, что с ростом числа комнат будет расти и цена.

### Определение самой высокой и самой низкой стоимости объекта

Определим в каких населенных пунктах из нашего набора данных располагаются квартиры с самой высокой и самой низкой стоимостью квадратного метра жилья.

Для полученных населенных пунктов с наибольшим количеством объявлений определим среднюю цену за квадратный метр.

```python
# сгруппируем данные по названию населенного
# пункта, считаем количество объявлений,
# находим средние значения для цены за кв.м.,
# сортируем значения по количеству объявлений,
# выбираем первые 10 и сортируем их по цене
df.groupby('locality_name'
          ).agg({
                'last_price': 'count', 
                'price_per_sqm': 'mean'}
            ).rename(columns={'last_price': 'count'}).sort_values(
                by='count', 
                ascending=False).head(10).sort_values(by='price_per_sqm',ascending=False)
```

Выделим факторы, которые влияют на стоимость квартиры в центре г. Санкт-Петербург.

```python
# построим графики зависимости цены квартиры 
# в центре С-П от параметров
for col in ['rooms', 'floor', 'cityCenters_nearest']:
    df_saint_petersburg_centre.pivot_table(
        index=col, 
        values='last_price', 
        aggfunc='median'
        ).plot(grid=True, 
               figsize=(12, 5), 
               title= f'{col} в центре Санкт-Петербурга')
```

Цена квартиры в центре Санкт-Петербурга растет с ростом количества комнат и увеличением этажа, на котором находится квартира. Оценим как фактор даты размещения объявления о продаже недвижимости влияет на цену.

```python
# сгруппируем данные по дню размещения 
# объявления, найдем медианное значение цены,
# построим столбчатый график
df_saint_petersburg_centre.pivot_table(
    index='exposition_day', 
    values='last_price', 
    aggfunc='median'
    ).sort_values(by='last_price'
    ).plot(kind='bar',
          grid=True, 
          figsize=(12, 5), 
          title='Зависимость цены от дня размещения в центре Санкт-Петербурга');
```

Здесь видим, что цена на объект недвижимости выше в объявлениях, опубликованных в воскресенье и ниже всего - в субботу.

```python
# сгруппируем данные по месяцу размещения 
# объявления, найдем медианное значение цены,
# построим столбчатый график
df_saint_petersburg_centre.pivot_table(
    index='exposition_month', 
    values='last_price', 
    aggfunc='median'
    ).sort_values(by='last_price'
    ).plot(kind='bar', 
          grid=True, 
          figsize=(14, 5),
          title='Зависимость цены от месяца размещения в центре Санкт-Петербурга');
```

Согласно графику, выше всего цена в объявлениях опубликованных в феврале и сентябре, а ниже всего - в декабре.

```python
df_saint_petersburg_centre.pivot_table(
    index='exposition_year', 
    values='last_price', 
    aggfunc='median'
    ).plot(kind='bar', 
          grid=True, 
          figsize=(14, 5),
          title='Зависимость цены от месяца размещения в центре Санкт-Петербурга');
```

Несомненно зависимость цены квартиры в центре Санкт-Петербурга от года размещения есть. Мы видим, что цены несколько снизились в 2015 году, но в дальнейшем с 2017 года мы наблюдаем на графике их рост.
